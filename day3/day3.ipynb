```python
import re
```

```python
with open("input3.txt") as f:
    instructions = f.read().replace("\n", "")
```

```python
instructions
```

```python
pattern = re.compile(r"mul\((\d{1,3},\d{1,3})\)")
```

```python
matched_instructions = [inst.split(",") for inst in re.findall(pattern, instructions)]
```

```python
print(matched_instructions)
```

```python
sum = 0
for a, b in matched_instructions:
    sum += int(a) * int(b)

print(sum)
```

```python
# Part 2
# The do() instruction enables future mul instructions.
# The don't() instruction disables future mul instructions.
```

```python
do_splits = re.split(r"do\(\)", instructions)
```

```python
print(do_splits)
```

```python
do_instructions = []

for inst in do_splits:
    # split at don't()
    without_dont = re.split(r"don\'t\(\)", inst)[0]
    do_instructions += [inst.split(",") for inst in re.findall(pattern, without_dont)]
```

```python
do_sum = 0
for a, b in do_instructions:
    do_sum += int(a) * int(b)

print(do_sum)
```
