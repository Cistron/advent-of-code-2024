```python
import numpy as np
import re
```

```python
# loading input
with open('input4.txt', 'r') as file:
    puzzle = file.readlines()
```

```python
# stripping any whitespaces
puzzle = [x.strip() for x in puzzle]
```

```python
puzzle_matrix = np.array([list(l) for l in puzzle])
```

```python
puzzle_matrix
```

```python
# create all diagonal lines from matrix
w, h = puzzle_matrix.shape
max_offset = w
min_offset = -h

diagonals1 = [puzzle_matrix.diagonal(i) for i in range(min_offset, max_offset)]
diagonals2 = [np.flipud(puzzle_matrix).diagonal(i) for i in range(min_offset, max_offset)]
```

```python
diagonals2
```

```python
# get vertical lines
verticals = [col for col in puzzle_matrix.T]
```

```python
verticals
```

```python
# adding diagonals to the puzzle
puzzle_w_diagonals_verticals = (
    puzzle + ["".join(d) for d in diagonals1] + ["".join(d) for d in diagonals2] + ["".join(d) for d in verticals]
)
```

```python
puzzle_w_diagonals_verticals
```

```python
pattern = re.compile(r"(?=(XMAS|SAMX))")

xmas_matches = 0
# searching for the pattern in the puzzle
for l in puzzle_w_diagonals_verticals:
    # xmas_matches += len(re.findall(pattern, l))
    matches = re.finditer(pattern, l)
    xmas_matches += len([match.group(1) for match in matches])

print(xmas_matches)
```

```python
# looking for the pattern in the puzzle
# it can be in any direction
# M.S
# .A.
# M.S

# what if I get all 3x3 squares and check if they contain the pattern?
# I can flatten the 3x3 square and write a regex that matches all possible combinations
```

```python
# get all 3x3 squares
squares = []
h, w = puzzle_matrix.shape

for i in range(0, w - 2):
    for j in range(0, h - 2):
        square = puzzle_matrix[i : i + 3, j : j + 3]
        squares.append(square)
```

```python
# variations of the pattern
# M.S
# .A.
# M.S

# S.M
# .A.
# S.M

# M.M
# .A.
# S.S

# S.S
# .A.
# M.M

pattern = re.compile(r"M.S.A.M.S|S.M.A.S.M|M.M.A.S.S|S.S.A.M.M")

mas_matches = 0

for sq in squares:
    flat_sq = "".join(sq.flatten())
    if pattern.match(flat_sq):
        mas_matches += 1

print(mas_matches)
```
