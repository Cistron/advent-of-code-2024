```python
# data = "2333133121414131402"

with open("input.txt") as f:
    data = f.read().strip()
```

```python
files = data[::2]
space = data[1::2]
```

```python
data_list = []
for i, blocksize in enumerate(files):
    if i < len(space):
        gap = int(space[i])
    else:
        gap = 0
    block = [str(i)] * int(blocksize) + ["."] * gap
    data_list += block
```

```python
print(data_list[-200:])
```

```python
# optimized filling of empty spaces using two-pointer approach
i, j = 0, len(data_list) - 1
while i < j:
    if data_list[i] == ".":
        while data_list[j] == "." and i < j:
            j -= 1
        if i < j:
            data_list[i], data_list[j] = data_list[j], data_list[i]
            j -= 1
    i += 1
```

```python
# calculate checksum

checksum = 0
for i, file_id in enumerate(data_list):
    if file_id != ".":
        checksum += int(file_id) * i

print(checksum)
```

```python
data = "2333133121414131402"

# with open("input.txt") as f:
#     data = f.read().strip()

files = data[::2]
space = data[1::2]

data_list = []
for i, blocksize in enumerate(files):
    file_id_block = (str(i), int(blocksize))
    data_list.append(file_id_block)
    if i < len(space):
        gap_size = int(space[i])
        if gap_size > 0:
            gap_block = (".", gap_size)
            data_list.append(gap_block)
```

```python
data_list_len = len(data_list)
i, j = 0, data_list_len - 1

while i < len(data) and j >= 0:
    # if a file block (iterating from back)
    if data_list[j][0] != ".":
        # searching for a non-file block
        while data_list[i][0] != "." and i < len(data):
            i += 1
        if data_list[i][0] == "." and data_list[j][1] == data_list[i][1] and i < len(data):
            data_list[i], data_list[j] = data_list[j], data_list[i]
            i += 1
        elif data_list[i][0] == "." and data_list[j][1] <= data_list[i][1] and i < len(data):
            file_block = data_list[j]
            gap_block = data_list[i]
            size_diff = gap_block[1] - file_block[1]
            # shrinking the gap block
            gap_block = (gap_block[0], gap_block[1] - size_diff)
            # swapping new block location
            data_list[j] = gap_block
            data_list[i] = file_block
            # adding extra gap block after new file block location
            data_list.insert(i+1, (".", size_diff))
            data_list_len += 1
            i += 2  # because an extra gap was added
    j -= 1
    data_string = ""
    for file_id, size in data_list:
        data_string += file_id * size
    print(data_string)
```
