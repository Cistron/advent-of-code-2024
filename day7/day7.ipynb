```python
import itertools
```

```python
with open("input.txt") as f:
    data = f.readlines()

data = [x.strip() for x in data]
data = [[int(y) for y in x.replace(":", "").split(" ")] for x in data]
```

```python
data[:5]
```

```python
def evaluate_calculation(lst: list[int]) -> int:
    # find all possible combinations of products and sums going left to right through the data
    # check if the result is the same as the expected result
    result = lst[0]
    start = lst[1]
    numbers = lst[2:]

    # create all possible operator combinations (cartesian product)
    # 0 = +
    # 1 = *
    operator_instructions = list(itertools.product([0, 1], repeat=len(numbers)))

    results = []
    for instruction in operator_instructions:
        total = start
        for n, op in zip(numbers, instruction):
            if op == 0:
                total += n
            else:
                total *= n
        results.append(total)
    
    if result in results:
        return result
    else:
        return 0
```

```python
total_calibration_results = []

for cal in data:
    total_calibration_results.append(evaluate_calculation(cal))

print(total_calibration_results)
```

```python
print(sum(total_calibration_results))
```

```python
def evaluate_calculation2(lst: list[int]) -> int:
    # find all possible combinations of products and sums going left to right through the data
    # check if the result is the same as the expected result
    result = lst[0]
    start = lst[1]
    numbers = lst[2:]

    # create all possible operator combinations (cartesian product)
    # 0 = +
    # 1 = *
    # 2 = concatenate
    operator_instructions = list(itertools.product([0, 1, 2], repeat=len(numbers)))

    for instruction in operator_instructions:
        total = start
        for n, op in zip(numbers, instruction):
            if op == 0:
                total += n
            elif op == 1:
                total *= n
            else:
                total = int(f"{total}{n}")
        if total == result:
            return result        

    return 0
```

```python
total_calibration_results2 = []

for cal in data:
    total_calibration_results2.append(evaluate_calculation2(cal))

print(sum(total_calibration_results2))
# 354060705047464
# 13.5s
# 6.6s with early return
```
